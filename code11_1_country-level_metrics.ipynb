# Country-level Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix price in different countries.csv')
print("", df.shape)
df.head(10)
```

```python
df1 = pd.read_csv('Netflix subscription fee Dec-2021.csv')
print("", df1.shape)
df1.head(10)
```

```python
gdp = pd.read_csv('GDP.csv',skiprows=4)
gdp['GDP'] = gdp.ffill(axis=1).iloc[:, -1] # select most recent available data
gdp = gdp[['Country Name', 'GDP']].replace('NY.GDP.MKTP.CD', np.nan)
gdp = gdp.rename(columns={'Country Name': 'Country'})
df = df.merge(gdp, how='left', on='Country')
```

```python
population = pd.read_csv('Population.csv',skiprows=4)
population['Population'] = population.ffill(axis=1).iloc[:, -1]
population = population[['Country Name','Population']]
population = population.rename(columns={'Country Name': 'Country'})
df = df.merge(population, how='left', on='Country')
```

```python
gini = pd.read_csv('Gini_Index.csv',skiprows=4)
gini['Gini_Index'] = gini.ffill(axis=1).iloc[:, -1]
gini = gini[['Country Name','Gini_Index']].replace('SI.POV.GINI', np.nan)
gini = gini.rename(columns={'Country Name': 'Country'})
df = df.merge(gini, how='left', on='Country')
```

```python
happiness = pd.read_csv('share-of-people-who-say-they-are-happy.csv')
happiness = happiness.groupby(['Entity'],as_index=False).last()
happiness = happiness.rename(columns={'Entity': 'Country', \
                                      'Share of people who are happy (World Value Survey 2014)': 'Happiness_Share'})
happiness = happiness[['Country','Happiness_Share']]
df = df.merge(happiness, how='left', on='Country')
```

```python
satisfaction = pd.read_csv('happiness-cantril-ladder.csv')
satisfaction = satisfaction.groupby(['Entity'],as_index=False).last()
satisfaction = satisfaction.rename(columns={'Entity': 'Country', \
                            'Life satisfaction in Cantril Ladder (World Happiness Report 2021)': 'Life_Satisfaction'})
satisfaction = satisfaction[['Country','Life_Satisfaction']]
df = df.merge(satisfaction, how='left', on='Country')
```

```python
broadband = pd.read_csv('Broadband_subscriptions.csv',skiprows=4)
broadband['Broadband_subscriptions'] = broadband.ffill(axis=1).iloc[:, -1]
broadband = broadband[['Country Name','Broadband_subscriptions']]
broadband = broadband.rename(columns={'Country Name': 'Country'})
df = df.merge(broadband, how='left', on='Country')
```

```python
cellular = pd.read_csv('Mobile_cellula_ subscriptions.csv',skiprows=4)
cellular['Cellular_subscriptions'] = cellular.ffill(axis=1).iloc[:, -1]
cellular = cellular[['Country Name','Cellular_subscriptions']]
cellular = cellular.rename(columns={'Country Name': 'Country'})
df = df.merge(cellular, how='left', on='Country')
```
