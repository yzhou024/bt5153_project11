```python
df_basics = pd.read_csv('title.basics.tsv',sep = '\t')
df_basics.head(10)
```

```python
# to simplify, only keep the week of 2021-11-14

df_top = pd.read_csv('all-weeks-countries.csv')
df_top = df_top[df_top.week == '2021-11-14']
df_top = df_top[df_top.country_name.isin(list_country)]

df_top.reset_index(inplace = True, drop = True)
df_top
```

```python
top_list = df_top.show_title.unique().tolist()
len(top_list)
```

```python
original_list = df_basics.originalTitle.tolist()
len(original_list)
```

```python
a = []
for i in top_list:
    if i in original_list:
        a.append(1)
    else:
        a.append(0)
```

```python
import collections

w = collections.Counter(a)
w
```

```python
primary_list = df_basics.primaryTitle.tolist()
len(primary_list)
```

```python
b = []
for i in top_list:
    if i in primary_list:
        b.append(1)
    else:
        b.append(0)

w = collections.Counter(b)
w
```

```python
len(b)
```

```python
c = []
for i in range(len(a)):
    if a[i] ==0  and b[i] == 0:
        #print(i)
        c.append(i)

len(c)
```

```python
df_new = df_top

for i in range(len(df_new)):
    print(i)
    title = df_new.iloc[i,5]
    if title in original_list or title in primary_list:
        print(str(title) + '=' + str(i))
        
        df_new.loc[i,'tconst'] = str(df_basics[df_basics.primaryTitle == title].tconst.iloc[0])

df_new
```
