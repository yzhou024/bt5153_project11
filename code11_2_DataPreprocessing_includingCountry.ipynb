```python
imdb_rating = pd.read_csv('title.ratings.tsv', sep='\t')
imdb_rating
```

```python
tv_episode = pd.read_csv('title.episode.tsv', sep='\t')
tv_episode
```

```python
tv_episode['tv_or_movie'] = 'tv' 
tv_episode
```

```python
imdb_rating = pd.merge(imdb_rating, tv_episode[['tconst', 'tv_or_movie']], on='tconst', how='left')
imdb_rating['tv_or_movie'] = imdb_rating['tv_or_movie'].fillna('movie')
imdb_rating
```

```python
imdb_rating[imdb_rating['numVotes'] > 100].copy()['tv_or_movie'].value_counts()
```

```python
rating_df = imdb_rating[imdb_rating['numVotes'] > 100].copy()
```

```python
df = pd.read_csv('Netflix subscription fee Dec-2021.csv')
df['Country_code'] = df['Country_code'].str.upper()
df.head(5)
```

```python
df.boxplot(column='No. of TV Shows', return_type='axes')
```

```python
df.boxplot(column='No. of Movies', return_type='axes')
```

```python
tv_ave_rating_list = []
movie_ave_rating_list = []
for i in list(df.index.values):
    tv_ave_rating_list.append(rating_df.sample(n = df.loc[i, 'No. of TV Shows'])['averageRating'].mean())
    movie_ave_rating_list.append(rating_df.sample(n = df.loc[i, 'No. of Movies'])['averageRating'].mean())
```
