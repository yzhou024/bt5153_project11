```python
import pandas as pd 
import numpy as np
```

```python
df_1 = pd.read_csv('raw_0419_late.csv')
df_2 = pd.read_csv('Top10Rating&VoteByCountry.csv')
```

```python
df_1
```

```python
df_2
```

```python
df=df_1.merge(df_2,how='left',left_on='Country',right_on='country_name')
df.drop('country_name',axis = 1, inplace=True)
df.fillna({'averageRating':df.averageRating.mean(), 'numVotes':df.numVotes.mean()}, inplace=True)

df
```

```python
df.describe()
```

```python
df.to_csv('data_with_top10_rating&votes.csv')
```

```python
df_test = df.sample(frac=0.2,random_state=5153)
df_test.to_csv('test.csv')
```

```python
df_test.index.tolist()
```

```python
list = []

for i in range(len(df)):
    if i not in df_test.index.tolist():
        list.append(i)
list
```
